Find min and max values of quantity and item price by using apply method

```python
#Write your code here
# sytax of lambda: lambda arguments: expression ex: add_ten = lambda x: x + 10
max_price = df.item_price.apply(lambda x: float(x[1:])).max()
min_price = df.item_price.apply(lambda x: float(x[1:])).min()
print(max_price, min_price)
```

We will write code to answer the following questions:

For question 1, we will use the apply method to create a column that will read "Burrito" for items that are a burrito, "Bowl" for items that are a bowl and "Other" for every other item.

```python
#Write your code here
# classify_number = lambda x: 'Positive' if x > 0 else ('Negative' if x < 0 else 'Zero')
df["Bowl_Burrito"] = df["item_name"].apply(lambda x: "Burrito" if "Burrito" in x else ("Bowl" if "Bowl" in x else "Other"))
# df.drop(columns=["new_column"], inplace =True)
df

```

Practice with apply function

Consider the dataframe for three baseball palyers below.

The apply() method is one of the most common methods of data preprocessing. It simplifies applying a function on each element in a pandas Series and each row or column in a pandas DataFrame. In this part, we'll learn how to use the apply() method in pandas — you'll need to know the fundamentals of Python and lambda functions. 


```python
import pandas as pd

#Create dataframe
names = ["Chapman, Matt", "Olson, Matt", "Davis, Khris"]
positions = ["3B", "1B", "OF"]
hrs = [25,30,45]
df_players = pd.DataFrame({"names":names, "position":positions, "homeruns":hrs})
df_players
```

Question 1. Use apply to add a column for each player's last name.

```python
#Write your code here
df_players["last_name"]=df_players.names.apply(lambda x : x.strip().split(" ")[1])
df_players
```

Question 2. Use apply to add a column for each player's first name.

```python
#Write your code here
df_players["first_name"]=df_players.names.apply(lambda x : x.strip().split(" ")[0][:-1])
df_players
```

Question 3. Add a column denoted whether the player is an all-star. We will say a player is an all-star if they play third base or outfield and have at least threshold_hr = 20 homeruns. If the player is an all-star the row in the new column should read "all-star" and otherwise it shoul read "not selected".

```python
#Write your code here
# df_players.apply(lambda row : "all-star" if (row[1] == "3B" or row[1] == "OF") and row[2] >= 20 else "not selected" \
#                  , axis =1)
df_players["all_star"]=df_players.apply(lambda row : "all-star" if (row[1] == "3B" or row[1] == "OF") and row[2] >= 20 else "not selected" \
                 , axis =1)
df_players
```
