How many missing entries are there in the Age column? Use the isnull() method and recall than Booleans are exactly like 1s (True) and 0s (False).

```python
#Write you code here
df2["Age"].isnull().sum()
```

Compute the avg age of passengers ignoring the missing data.

```python
#Write your code here
# By default, the mean() method in pandas ignores NaN values. So we don't have to add .dropna()
df2["Age"].mean()
# df2["Age"].dropna().mean()
```

Using the fillna() method replace the missing values in the Age column with the mean.

```python
#Write your code here
df2['Age'].fillna(df2['Age'].mean(), inplace=True)
df2
```

What is the average age of the 5 oldest passengers? The reset_index method will be helpful here.

```python
#Write your code here
df2.nlargest(5,"Age")["Age"].mean()
```

```python
# Sort the DataFrame by the 'Age' column in descending order
df_titanic_sorted = df2.sort_values(by='Age', ascending=False)
df_titanic_sorted
# Reset the index of the sorted DataFrame
df_titanic_sorted.reset_index(drop=True, inplace=True)
df_titanic_sorted
# Select the top 5 oldest passengers and calculate the average of their ages
avg_age_oldest_5 = df_titanic_sorted['Age'].head(5).mean()

print(avg_age_oldest_5)
```
